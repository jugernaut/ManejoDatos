## Tipos de Simulaciones

Actualmente existen bsicamente 2 tipos de simulaciones. La primera de ellas son las simulaciones basadas en **modelos matemticos complejos** que incorporan el uso de ecuaciones diferenciales, sistemas de ecuaciones o integrales para describir el comportamiento de un fenmeno, este tipo de simulaciones se estudian en materias como **anlisis numrico** o **modelacin matemtica**.

El otro tipo de simulaciones son las **simulaciones basadas en agentes** (antes autmatas celulares). En este tipo de simulaciones, la idea principal es tratar de determinar las **interacciones locale entre los individuos** que conforman el sistema, para poder determinar cual sera el comportamiento del sistema en el paso siguiente de tiempo.

Sin importar el tipo de simulacin la forma en la que esta debe ser construida no cambia mucho.

El algoritmo general seria el siguiente. 

1. Se observa un fenmeno de inters.
2. Se genera un modelo matemtico (ya sea complejo o basado en agentes).
3. Se construye el correspondiente modelo computacional.
4. Finalmente se procede a calibrar y probar la simulacin.

Vale la pena aclarar que siempre que sea posible obtener la solucin del problema planteado, de manera analtica es preferible seguir este camino, ya que es el camino que mayor exactitud y confiabilidad ofrece.

# Autmatas Celulares (modelos basados en agentes)

Un autmata celular es un modelo matemtico que, basado en las interacciones locales permite mostrar la evolucin de un sistema dinmico que avanza a pasos discretos de tiempo. 

Existe diferentes cientficos que han aportado en gran medida a lo que hoy se conoce como autmata celular (A.C.), sin embargo se puede decir que la persona responsable de la definicin de A.C. fue **John von Newmann**.

Por otro lado **Stephen Wolfram** llevo a un nuevo nivel la definicin de A.C. ya que el los clasifico y realizo mltiples investigaciones muy importantes al respecto.

Dicho sea de paso Stephen Wolfram es el creador de **Wolfram Alpha**, gran aliado y salvador de todo estudiante de la Facultad de Ciencias. 

## Descripcin

No se tiene como tal una definicin matemtica formal de lo que es un A.C., pero podemos pensar en el mismo como una tupla, es decir un conjunto de elementos ordenados

* Una rejilla o cuadricula (lattice) de enteros, infinitamente extendida y con dimensin $d\in\mathbb{Z^{+}}$. Cada celda es conocida como **clula**.

* Una clula puede tomar un valor en $\mathbb{Z}$ a partir de un **conjunto finito de estados $k$**.

* Cada clula ademas se caracteriza por su **vecindad**, conjunto finito de clulas que colindan con la clula en cuestin.

* La **funcin de transicin**, toma el estado actual de la clula y los valores de la vecindad de la clula en cuestin y regresa el nuevo valor de la clula que tendr en el paso siguiente de tiempo. Esta funcin se aplica de manera homognea a todas las clulas del A.C. por cada paso de tiempo.

# Aproximacin de $\pi$ (Montecarlo)


El ejemplo ms conocido para mostrar como funcionan los mtodos de Monte Carlo, es el de la aproximacin del valor de $\pi$.

Existen diferentes versiones de como fue ideada esta aproximacin pero normalmente se encuentra en la literatura como aproximacin de $\pi$ mediante lanzamiento de dardos o aproximacin de $\pi$ empleando gotas de lluvia.

La idea bsica de esta aproximacin es generar pares de nmeros aleatorios dentro de un dominio y calcular la probabilidad de que estos nmeros aleatorios pertenezcan a un subconjunto del dominio, es decir.

Consideremos la circunferencia unitaria, circunscrita dentro de un cuadrado de lado 2.

Ahora supongamos que alguien comienza a lanzar dardos que caen dentro del cuadrado y ademas estn **distribuidos de manera uniforme**.

El marco terico de los mtodos de Monte Carlo indica que podemos aproximar el valor de $\pi$ a partir de calcular la probabilidad de que los dardos lanzados caigan dentro de la circunferencia unitaria circunscrita dentro del cuadrado de lado 2. 

De manera tal que $P_{dardos}$ sera la probabilidad de que los dardos caigan dentro de la circunferencia y para calcular esta probabilidad basta con realizar el siguiente calculo. 

$$P_{dardos}=\frac{Area\,Circunferencia}{Area\,Cauadrado}$$

Recordando un poco de clculo.

$$\frac{Area\,Circunferencia}{Area\,Cauadrado}=\frac{\int_{-1}^{1}dx\int_{-\sqrt{1-x^{2}}}^{\sqrt{1-x^{2}}}dy}{\int_{-1}^{1}dx\int_{-1}^{1}dy}=\frac{2\int_{-1}^{1}\sqrt{1-x^{2}}dx}{2*2}=\frac{\pi}{4}$$

De lo que podemos concluir que

$$P_{dardos}=\frac{Area\,Circunferencia}{Area\,Cauadrado}=\frac{\pi}{4}$$

Es decir que

$$ \pi=4*P_{dardos}$$

Finalmente solo resta dividir entre el numero de lanzamientos que se hayan realizado para obtener una aproximacin del valor de $\pi$.

$$\pi\approx\frac{4*P_{dardos}}{N}$$


Donde $N$ es la cantidad de lanzamientos de dardos que se hayan realizado y con lo mencionado previamente, a mayor numero de lanzamientos (experimentos, muestras, etc.) la aproximacin encontrada ser ms cercana al valor real.

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.html.widgets import interact
import ipywidgets as widgets

def mc_pi_aprox(n):
    # tamano de la imagen
    plt.figure(figsize=(10,10))
    # se generan valores de x,y de manera uniforme (SIMULACION DE PROBABILIDAD)
    x, y = np.random.uniform(-1, 1, size=(2, n))
    # interior contiene la cantidad de dardos
    # que calleron en la circunferencia (Pdardos)
    interior = (x**2 + y**2) <= 1
    # la suma de interior indica cuantas coordenadas estan en la circunferencia
    Pdardos = interior.sum()
    # se calcula el valor aproximado de pi
    pi = Pdardos * 4 / float(n)
    # se calcula el error abs
    error = abs((pi - np.pi) / pi) * 100
    # dardos que calleron fuera
    exterior = np.invert(interior)
    # seccion de graficas
    plt.plot(x[interior], y[interior], 'b.')
    plt.plot(x[exterior], y[exterior], 'r.')
    plt.plot(0, 0, label='$\hat \pi$ = {:4.4f}\nerror = {:4.4f}%'
             .format(pi,error), alpha=0)
    plt.legend(frameon=True, framealpha=0.9, fontsize=16)

interact(mc_pi_aprox,n=widgets.IntSlider(min=0,max=10000,step=1000,value=10))
```
